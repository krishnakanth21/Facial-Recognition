```python
import math
videoFile = "sample2.mp4"
imagesFolder = "C:\\Users\\super\\OneDrive\\Desktop\\Images"
cap = cv2.VideoCapture(videoFile)
frameRate = cap.get(5) #frame rate
while(cap.isOpened()):
    frameId = cap.get(1) #current frame number
    ret, frame = cap.read()
    if (ret != True):
        break
    if (frameId % math.floor(frameRate) == 0):
        filename = imagesFolder + "/image_" +  str(int(frameId)) + ".jpg"
        cv2.imwrite(filename, frame)
cap.release()
print ("Done!")
```

```python
haar_cascade_face = cv2.CascadeClassifier('C:\\Users\\super\\Face-Detection-in-Python-using-OpenCV-master\\data\\haarcascades\\haarcascade_frontalface_default.xml')
```

```python
def detect_faces(cascade, test_image, scaleFactor = 1.2):
    image_copy = test_image.copy()
    
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
  
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=5)
    
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)
        
    return image_copy
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```
